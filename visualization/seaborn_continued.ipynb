```python
import seaborn as sns
```

```python
#import dataset
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
#strip plot usign axes level function
sns.stripplot(tips,x='day',y='total_bill')
```

```python
#using catplot
#figure level function 
sns.catplot(tips,x='day',y='total_bill',kind='strip')
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='strip',jitter=0.1,hue='sex')
```

```python
#swarmplot
#figure level
sns.catplot(tips,x='day',y='total_bill',kind='swarm')
```

```python
#axes level
sns.swarmplot(tips,x='day',y='total_bill',hue='sex')
```

```python
#box plot
sns.boxplot(tips,x='day',y='total_bill')
```

```python
#figure level
sns.catplot(tips,x='day',y='total_bill',kind='box')
```

```python
#hue
sns.catplot(tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
#single boxplot -> numerical column
sns.boxplot(tips,y='total_bill')
```

Violinplot = (Boxplot + KDE plot)

```python
#violin plot
sns.violinplot(tips,x='day',y='total_bill')
```

```python
#using catplot -> feature level
sns.catplot(tips,x='day',y='total_bill',kind='violin')
```

```python
#hue
sns.catplot(tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
#bar plot
sns.barplot(tips,x='sex',y='total_bill',)
```

```python
import numpy as np
sns.catplot(tips,x='sex',y='total_bill',kind='bar',errorbar=None,hue='smoker',estimator=np.max)
```

```python
#point plot
sns.pointplot(tips,x='sex',y='total_bill',estimator='mean',hue='smoker',errorbar=None)
```

```python
#countplot
sns.countplot(tips,x='sex',hue='day')
```

```python
#faceting
sns.catplot(tips,x='sex',y='total_bill',kind='box',col='smoker',row='time')
```
