```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #to use its dataset

plt.style.use('default')
```

```python
#plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price) #2Dline plot
```

```python
#from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
print(batsman)

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Seanson')
plt.ylabel('Runs Scored')
```

```python
#color(hex) and line(width and style) and marker(size) 
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#941982')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Seanson')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid', linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#941982',linestyle='dashdot', linewidth=2)

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Seanson')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid', linewidth=3, marker='.', markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#941982',linestyle='dashdot', linewidth=2, marker='o')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Seanson')
plt.ylabel('Runs Scored')
```

```python
#legend -> location
#color(hex) and line(width and style) and marker(size) 
plt.plot(batsman['index'],batsman['V Kohli'],color='green',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#941982',label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Seanson')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
#color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
#multiple Bar charts
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='red')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
 #Stacked bar chart
df
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')

plt.legend()
plt.show()
```

```python
#dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
#percnetage and colors
plt.pie(df['batsman_runs'], labels=df['batsman'],autopct='%0.1f%%',colors=['red','blue','green','pink','yellow','cyan'])
plt.show()
```

```python
#explode shadow
plt.pie(df['batsman_runs'], labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
