**Relational plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #to import dataset it contains 
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#scatter plot -> axes level function, added hue, style, size
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
#relplot -> figure level -> square shape
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
#Line plot
gap = px.data.gapminder()
temp_df = gap[gap['country']=='India']
temp_df
```

```python
#axes level function
sns.lineplot(temp_df,x='year',y='lifeExp')
```

```python
#using relplot
sns.relplot(temp_df,x='year',y='lifeExp',kind='line')
```

```python
#hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(temp_df,x='year',y='lifeExp',kind='line',hue='country',size='continent')
```

```python
#facet plot -> only with figure level functions ->work with relplot
#it will not work with scatterplot and lineplot
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
sns.relplot(gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

**Distribution PLot**

```python
#axes level →histplot() → kdeplot() → rugplot
#figure level → displot()
```

```python
#plotting univariate histogram
sns.histplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='hist')
```

```python
#bin size
sns.displot(tips,x='total_bill',kind='hist',bins=2)
```

```python
#It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
#Discrete bins are automatically set for categorical variables

#countplot
sns.displot(tips,x='day',kind='hist')
```

```python
#hue parameter
sns.displot(tips,x='tip',kind='hist',hue='sex')
```

```python
#element -> step
sns.displot(tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(titanic,x='age',kind='hist',hue='sex',element='step')
```

```python
#faceting using col and row -> not work on histplot function
sns.displot(tips,x='total_bill',kind='hist',col='sex')
```

```python
#kde plot
#Rather than using discrete bins, a KDE plot smooths the observations with a Guassian, producing a continuous density estimate
sns.kdeplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='kde')
```

```python
#hue parameter
#plt.figure(figsize=(15,15)) -> not work on figure level
sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=1)
```

```python
#Rug plot
# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
#bivariate histogram
#A bivariate histogram bins the data within rectangles that title the plot
#and then shows the count of observations within each rectangle with the fill color
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

**Matrix Plot**

```python
#heatmap
#Heat map plots a rectangular data as a color-encoded matrx

temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
#customization with heatmp
```

```python
#annot, linewidth, cmap,
temp_df = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='viridis')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
