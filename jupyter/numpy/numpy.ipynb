**Be careful when copying array!!!**

```python
a = np.array([1,2,3])
b = a
print(b)
b[0] = 100
print(b)
print(a)

#To prevent this 
a = np.array([1,2,3])
b = a.copy()
print(b)
b[0] = 100
print(b)
print(a)
```

**Reorganizing Arrays**

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after = before.reshape((4,2))
print(after)
```

```python
after = before.reshape((2,2,2)) #as long as it has same number of values
print(after)
```

```python
#Vertically stacking vectors 
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
```

```python
#Horizontal stacking vectors
np.hstack([v1,v2])
```

```python
h1 = np.ones((2,3))
h2 = np.zeros((2,2))
np.hstack((h1,h2))
```
